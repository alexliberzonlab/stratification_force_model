```python
import numpy as np
# from scipy.io import loadmat
import matplotlib.pyplot as plt
# import flowtracks
import h5py
# import hdf5storage
import os
import glob
# from flowtracks.trajectory import Trajectory
from scipy.io import loadmat
```

```python
file_list = sorted(glob.glob('../Data/*.mat'))
print(file_list)
```

```python
for fname in file_list:
    data_name = fname.rsplit('/')[-1].split('_')[0]
    data = loadmat(fname)
    data = data[data_name]
    fig,ax = plt.subplots(1,2,figsize=(8,8))
    n_particles = data['time'].shape[1]
    for i in range(n_particles):
        ax[0].plot(data['z'][0][i].flatten(), data['vz'][0][i].flatten())
        ax[1].plot(data['time'][0][i].flatten(), data['z'][0][i].flatten())
        ax[0].set_title(data_name)
```
